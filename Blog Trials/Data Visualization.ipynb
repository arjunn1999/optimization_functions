```python
#imports  

from numpy import arange #The arange() function is used to get evenly spaced values within a given interval.
from numpy import meshgrid
from matplotlib import pyplot 
from mpl_toolkits.mplot3d import Axes3D
```

# UNIMODAL FUNCTIONS
    Funtions with single global optima (minima or maxima).Three examples of the functions are given below

Eg1: $$ z = x^2 + y^2 $$

```python

objective = lambda x,y :(x**2.0)+(y**2.0)
```

```python
r_min,r_max  = -5.0,5.0
x_axis = arange(r_min,r_max,0.1)
y_axis = arange(r_min,r_max,0.1)
```

```python
x,y=meshgrid(x_axis,y_axis)
```

```python
results = objective(x,y)
```

```python
figure = pyplot.figure()
ax  = pyplot.axes(projection='3d')
ax.plot_surface(x, y, results, cmap='jet')
pyplot.show()
```
